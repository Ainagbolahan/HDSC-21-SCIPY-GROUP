```python
#importing the neccessary libraries
import numpy as np
import pandas as pd
```

```python
#loading the dataset from github
url = 'https://github.com/Ainagbolahan/HDSC-21-SCIPY-GROUP/blob/2bf3af06838feee6e723c062afc6aa3fc36f2b31/dataset/Wurld2016_2021.csv?raw=true'
df = pd.read_csv(url)
```

```python
#checking how the data looks like
df.head()
```

```python
df.isnull().sum()
```

```python
df.tail()
```

```python
#checking for the columns with numerical values i.e columns with numerical dtypes
df.dtypes
```

```python
# Rank_y is a redundant column, so it is dropped
df.pop('Rank_y')
```

```python
#splitting the female and male ratio and creating a new dataframe
split = df['Female:Male Ratio'].str.split(':', n =1, expand =True)
new_df = pd.DataFrame({'Female %': split[0], 'Male %': split[1]})
#joining the dataframe to the original dataset
df = df.join(new_df)
```

```python
#dropping the female:male ratio column as it is no longer needed
df.pop('Female:Male Ratio')
```

```python
df.head()
```

```python
df.dtypes
```

```python
#changing the data type of the object columns to numerical data
df['No. of FTE Students'].replace('(,)', value = '', regex = True, inplace = True)
df['No. of FTE Students'] = pd.to_numeric(df['No. of FTE Students'], errors = 'coerce')
df['International Students'].replace('(%)', '', regex = True, inplace = True)
df['International Students'] = pd.to_numeric(df['International Students'], errors = 'ignore')
df['Overall'].replace('-', value = '', regex = True, inplace = True)
df['Overall'] = pd.to_numeric(df['Overall'], errors = 'coerce')
df['Industry Income'].replace('-', np.nan, regex = True, inplace = True)
df['Industry Income'] = pd.to_numeric(df['Industry Income'], errors = 'coerce')
df['Female %'] = pd.to_numeric(df['Female %'] , errors = 'coerce')
df['Male %'] = pd.to_numeric(df['Male %'], errors = 'coerce')
```

```python
df.head()
```

```python
df.dtypes
```

```python
#checking for null values
df.isnull().sum()
```

```python
df.pop('Overall')
```

```python
#filling in the missing data in the Industry Income column
df['Industry Income'].fillna(float(df['Industry Income'].mean()), inplace=True)
```

```python
#filling in the missing data in the Female% column
df['Female %'].fillna(round(int(df['Female %'].mean()), 0), inplace=True) 
```

```python
#filling in the missing data in the Male% column
df['Male %'].fillna(round(int(df['Male %'].mean()), 0), inplace=True)
```
